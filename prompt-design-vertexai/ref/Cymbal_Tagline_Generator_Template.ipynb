```python
from google import genai
from google.genai import types
import base64
import os

def generate():
  client = genai.Client(
      vertexai=True,
      api_key=os.environ.get("GOOGLE_CLOUD_API_KEY"),
  )

  msg1_text1 = types.Part.from_text(text="""Write a tagline for a durable backpack designed for hikers that makes them feel prepared. Consider styles like minimalist.""")
  msg2_text1 = types.Part.from_text(text="""**Querying for Inspiration**

I'm currently crafting a search query to uncover existing taglines for rugged hiking backpacks. The goal is to identify those that highlight preparedness, yet maintain a minimalist aesthetic. My hope is to glean some inspiration from the current market, or at least identify recurring themes and word choices.


**Examining Tagline Concepts**

I've been sifting through search results. My focus is on strong verbs and evocative nouns. I am currently concentrating on words that link durability with preparedness. I've noted down combinations like \"Built to Conquer\" and \"Ready for the Range.\"


**Synthesizing Tagline Ideas**

I've been analyzing the keywords and phrases. I'm focusing on blending the minimalist aspect with preparedness. I'm experimenting with a few more direct taglines, like \"Adventure, Simplified\" and \"Prepared, Unburdened.\" I'm also thinking about more conceptual ideas to enhance impact.


**Developing Core Taglines**

I'm now prioritizing brevity and impact, weaving in \"durable,\" \"prepared,\" and \"minimalist.\" I'm also considering the user's need for a minimalist theme. Current contenders: \"Built. Prepared. Explore.\" and \"Carry Less. Conquer More.\" I will narrow them down soon.""")
  msg2_text2 = types.Part.from_text(text="""Here are a few taglines for a durable, minimalist hiking backpack, designed to make hikers feel prepared:

*   **Trail Ready. Always.**
*   **Prepared. Simplified.**
*   **Essential Gear. Endless Trails.**
*   **Go Light. Go Far. Be Ready.**
*   **Your Preparedness, Refined.**""")
  msg4_text1 = types.Part.from_text(text="""**Assessing Tagline Concepts**

I've taken a look at the provided tagline and I'm focusing on similar structures. My current train of thought is to brainstorm a few alternatives, keeping the core concept but playing with different wording and emphasizing various benefits. I'm exploring ways to make it more concise and memorable.


**Analyzing Similar Structures**

I'm now zeroing in on the user's tagline and sketching out some alternative versions. I'm keen to preserve the essence of \"built for the journey\" and the adventure theme. My focus is now on achieving a more concise and memorable phrasing, emphasizing durability and preparedness.""")
  msg4_text2 = types.Part.from_text(text="""That's a strong tagline! It's clear, inspiring, and hints at durability and preparedness.

Here are a few more in a similar vein, playing on the themes of \"journey,\" \"essentials,\" and feeling prepared:

*   **Built for the Journey. Ready for Anything.**
*   **Your Journey's Essential. Always Prepared.**
*   **Every Journey, Mastered. With Your Essential Pack.**
*   **The Journey Awaits. Your Essentials Are Ready.**""")
  msg6_text1 = types.Part.from_text(text="""**Generating Cardigan Taglines**

I've begun brainstorming taglines, focusing on warmth and timeless appeal. I'm exploring phrases that emphasize the cardigan's durability for families and the connection it fosters. Currently, I'm leaning toward taglines evoking classic style and family bonding.""")
  msg6_text2 = types.Part.from_text(text="""Here are a few taglines for a durable cardigan designed for families, emphasizing a classic style and the idea of connection:

*   **Classic Comfort. Family Connected.**
*   **Woven Memories. Enduring Style.**
*   **Generations of Warmth. Always Together.**
*   **The Family Cardigan. Timeless Bonds, Lasting Comfort.**
*   **Connect. Comfort. Classic.**""")
  msg8_text1 = types.Part.from_text(text="""Here are some taglines that capture a \"fun feeling\" and the spirit of \"go live,\" perfect for encouraging young people to explore the outdoors:

*   **Go Live. Find Your Fun.**
*   **Unlock Adventure. Feel the Fun.**
*   **Your World Awaits. Live It. Love It.**
*   **Spontaneous Fun. Outdoor Bound.**
*   **Adventure Mode: ON. Fun Starts Now.**
*   **The Wild is Calling. Go Have Fun.**
*   **Live Bold. Play Free.**""")
  msg9_text1 = types.Part.from_text(text="""generate random interesting tagline base on following attributes:
- Product attributes: {{product_attributes}} - Target audience: {{audience}} - Emotional resonance: {{emotion}}""")
  msg10_text1 = types.Part.from_text(text="""Please provide specific values for `{{product_attributes}}`, `{{audience}}`, and `{{emotion}}` so I can generate interesting taglines for you!

For example, you could say:

**Product attributes:** Ultra-bright headlamp, long-lasting battery, waterproof
**Target audience:** Night hikers, stargazers, camp explorers
**Emotional resonance:** Discover, awe, security

Once you give me those details, I can create some taglines!""")
  si_text1 = """Cymbal Direct is partnering with an outdoor gear retailer. They're launching a new line of products designed to encourage young people to explore the outdoors. Help them create catchy taglines for this product line."""

  model = "gemini-2.5-flash"
  contents = [
    types.Content(
      role="user",
      parts=[
        msg1_text1
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg2_text1,
        msg2_text2
      ]
    ),
    types.Content(
      role="user",
      parts=[
        types.Part.from_text(text="""Built for the Journey: Your Adventure Essentials.""")
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg4_text1,
        msg4_text2
      ]
    ),
    types.Content(
      role="user",
      parts=[
        types.Part.from_text(text="""Write a tagline for a durable cardigan designed for families. Style: classic and connected.""")
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg6_text1,
        msg6_text2
      ]
    ),
    types.Content(
      role="user",
      parts=[
        types.Part.from_text(text="""fun feeling. go live""")
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg8_text1
      ]
    ),
    types.Content(
      role="user",
      parts=[
        msg9_text1
      ]
    ),
    types.Content(
      role="model",
      parts=[
        msg10_text1
      ]
    ),
    types.Content(
      role="user",
      parts=[
        types.Part.from_text(text="""product_attributes: durable, waterproof audience: families emotion: softness""")
      ]
    ),
  ]
  tools = [
    types.Tool(google_search=types.GoogleSearch()),
  ]

  generate_content_config = types.GenerateContentConfig(
    temperature = 1,
    top_p = 1,
    max_output_tokens = 65535,
    safety_settings = [types.SafetySetting(
      category="HARM_CATEGORY_HATE_SPEECH",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_DANGEROUS_CONTENT",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_SEXUALLY_EXPLICIT",
      threshold="OFF"
    ),types.SafetySetting(
      category="HARM_CATEGORY_HARASSMENT",
      threshold="OFF"
    )],
    tools = tools,
    system_instruction=[types.Part.from_text(text=si_text1)],
    thinking_config=types.ThinkingConfig(
      thinking_budget=-1,
    ),
  )

  for chunk in client.models.generate_content_stream(
    model = model,
    contents = contents,
    config = generate_content_config,
    ):
    if not chunk.candidates or not chunk.candidates[0].content or not chunk.candidates[0].content.parts:
        continue
    print(chunk.text, end="")

generate()
```
